```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
caminho = f'test.csv' ##Definindo o caminho do arquivo CSV
dados = pd.read_csv(caminho, sep=',') ## Abrindo o arquivo em um DataFrame
```

```python
# Plotando os dados:
plt.scatter(dados.x, dados.y)
plt.show()
```

```python
#Definindo a função de Gradiente Descendente
def gradient_descent(m_now, b_now, points, L):
    m_gradient = 0
    b_gradient = 0
    
    n = len(points)
    
    for i in range(n):
        X = points.iloc[i].x
        Y = points.iloc[i].y
        
        m_gradient += -(2/n) * X * (Y -(m_now * X + b_now))
        b_gradient  += -(2/n) * (Y -(m_now * X + b_now))
    
    m = m_now - m_gradient * L
    b = b_now - b_gradient * L
    return m, b

m = 0
b = 0
L = 0.0001 #Learning Rate da Regressão
epochs = 5000 #Quantidade de Iterações da Regressão 

for i in range(epochs): #Treinando a Regressão Linear
    m, b = gradient_descent(m,b,dados,L)
    if i % 50 ==0:
        print(f"Epoch: {i}, m: {m} e b: {b}")
print(m,b)

plt.scatter(dados.x, dados.y, color="black")
plt.plot(list(range(0,100)), [m * x + b for x in range(0,100)], color="red")
```
